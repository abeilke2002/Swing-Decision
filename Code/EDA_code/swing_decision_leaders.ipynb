```python
import pandas as pd
```

```python
take_file = '/Users/aidanbeilke/Desktop/Baseball Projects/Swing_Decision/take_leaders.csv'
swing_file = '/Users/aidanbeilke/Desktop/Baseball Projects/Swing_Decision/swing_leaders.csv'
```

```python
take = pd.read_csv(take_file)
swing = pd.read_csv(swing_file)
```

```python
take.head()
```

```python
swing.head()
```

```python
full = pd.merge(take, swing, on = 'full_name', how = 'inner')
```

```python
full = full[['full_name', 'take_rv', 'swing_rv']]

full.head()
```

```python
full['swing_decision_rv'] = full[['take_rv', 'swing_rv']].mean(axis=1)

full.head()
```

```python
final = full.sort_values(by = 'swing_decision_rv', ascending=False)

final
```

```python
final.to_csv('swing_decision_leaders.csv')
```
